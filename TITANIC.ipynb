# **Data Cleaning / Preprocessing**

> Data Cleaning means the process of identifying the incorrect, incomplete, inaccurate, irrelevant or missing part of the data and then modifying, replacing or deleting them according to the necessity. Data cleaning is considered a foundational element of the basic data science.

```python
df.columns.values
```

```python
miss_val=df.isnull().sum()
print(miss_val[miss_val>0])
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

* Most of the Age data has been missing. We need to fill this missing data instead of removing.

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=df,palette=["grey","Aquamarine","skyblue"]) 
```

* The above plot shows that Age depends on the class.

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age

df['Age']=df[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

* The missing values of Age has removed.

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
def impute_embarked(cols):
    Embarked = cols[0]
    
    if pd.isnull(Embarked):
        return 'S'
    else:
        return Embarked

df['Embarked']=df[['Embarked']].apply(impute_embarked,axis=1)
```

```python
df['Embarked'].isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

* We don't have any use of Cabin in our model so we are removed that Cabin.

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
pd.get_dummies(df['Embarked'],drop_first=True).head()
```

```python
sex = pd.get_dummies(df['Sex'],drop_first=True)
embark = pd.get_dummies(df['Embarked'],drop_first=True)
pclass = pd.get_dummies(df['Pclass'],drop_first=True)
```

* The dummy values (categorical values) in a column are replaced by binary digits using the method `.get_dummies` and the first columns are dropped.

```python
df.drop(['Sex','Embarked','Pclass','Name','Ticket'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df = pd.concat([df,sex,embark,pclass],axis=1)
```

```python
df.head()
```

* The above Data Frame contains cleaned data without any null values and are not of string dtype.

# **Splitting Data for Training and Testing**

```python
y = df["Survived"]
X = df.drop("Survived",axis=1)
```

* In this program we are assigning target column Survive to y and remaining data frame without Survived column to X.

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size = 0.2,random_state=1)
```

* The above program split the available data into 80% for training and 20% for testing.

# **Model Validation**

### Importing the required libraries/modules for finding the accuracy

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

# **Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
random_model = RandomForestClassifier(random_state=101)
```

> **Fitting** `.fit()` method is used for fitting a model

```python
random_model.fit(train_X,train_y)
```

* The model is fitted with train_X and train_y for further prediction.



> **Prediction** `.predict()` method is used for predicting the accuray i.e., the performance of model

```python
random_preds = random_model.predict(test_X)
```

* The predictions are made and are assigned to random_preds. Now the accuracy is to be calculated using the imported libraries/modules.

> **Score Calculation**

```python
print(classification_report(test_y,random_preds))
random_acc=round(metrics.accuracy_score(test_y,random_preds)*100,0)
print("Accuracy = ",random_acc)
```

# **Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
decision_model=DecisionTreeClassifier(random_state=10)
```

> **Fitting** `.fit()` method is used for fitting a model

```python
decision_model.fit(train_X, train_y)
```

* The model is fitted with train_X and train_y for further prediction.

> **Prediction** `.predict()` method is used for predicting the accuray i.e., the performance of model

```python
decision_preds =decision_model.predict(test_X)
```

> **Score Calculation**

```python
print(classification_report(test_y,decision_preds))
decision_acc=round(metrics.accuracy_score(test_y,decision_preds)*100,2)
print("Accuracy = ",decision_acc)
```
